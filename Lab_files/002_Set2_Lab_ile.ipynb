# 1.A program that models a bank account, with classes for the account, the customer, and the bank.

```python
class Customer:
    def __init__(self, name, address):
        self.name = name
        self.address = address

class Account:
    def __init__(self, account_number, customer, balance=0.0):
        self.account_number = account_number
        self.customer = customer
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return f"Deposited ${amount}. Current balance: ${self.balance}"

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"Withdrew ${amount}. Current balance: ${self.balance}"
        else:
            return "Insufficient funds."

class Bank:
    def __init__(self, name):
        self.name = name
        self.accounts = []

    def add_account(self, account):
        self.accounts.append(account)
        return f"Account {account.account_number} added to {self.name}'s bank."

# Example usage:
customer1 = Customer("Guru patil", "123 Main St")
account1 = Account(1001, customer1, 500.0)

customer2 = Customer("Mounesh Patil", "456 Oak St")
account2 = Account(1002, customer2, 1000.0)

my_bank = Bank("Sample Bank")
my_bank.add_account(account1)
my_bank.add_account(account2)

print(account1.deposit(200))
print(account2.withdraw(300))
```

# 2.A program that simulates a school management system, with classes for the students, the teachers, and the courses.

```python
class Student:
    def __init__(self, student_id, name, grade):
        self.student_id = student_id
        self.name = name
        self.grade = grade

class Teacher:
    def __init__(self, teacher_id, name, subject):
        self.teacher_id = teacher_id
        self.name = name
        self.subject = subject

class Course:
    def __init__(self, course_id, name, teacher, students=[]):
        self.course_id = course_id
        self.name = name
        self.teacher = teacher
        self.students = students

    def enroll_student(self, student):
        self.students.append(student)
        return f"{student.name} enrolled in {self.name}."

    def display_students(self):
        student_names = [student.name for student in self.students]
        return f"Students in {self.name}: {', '.join(student_names)}"

# Example usage:
student1 = Student(1, "Mahesh", 10)
student2 = Student(2, "Ganesh", 11)

teacher1 = Teacher(101, "Mr. Guru", "Python")
teacher2 = Teacher(102, "Ms. Johnson", "Math")

math_course = Course(1001, "Python 101", teacher1, [student1])
history_course = Course(1002, "Math 101", teacher2, [student2])

print(math_course.enroll_student(student2))
print(history_course.display_students())
```

# 3.A program that reads a text file and counts the number of words in it.

```python
def count_words(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            words = content.split()
            word_count = len(words)
            return word_count
    except FileNotFoundError:
        return "File not found."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage:
file_path = 'sample_text.txt'  # Replace with the path to your text file
result = count_words(file_path)

if isinstance(result, int):
    print(f"Number of words in the file: {result}")
else:
    print(result)
```

# 4. A program that reads a CSV file and calculates the average of the values in a specified column

```python
import pandas as pd

def calculate_column_average(file_path, column_name):
    try:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(file_path)

        # Check if the specified column name is present in the DataFrame
        if column_name not in df.columns:
            return f"Column '{column_name}' not found in the CSV file."

        # Calculate the average of the specified column
        average = df[column_name].mean()
        return average

    except FileNotFoundError:
        return "File not found."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage:
file_path = 'sample_data.csv'  # Replace with the path to your CSV file
column_name = 'column_name'  # Replace with the name of the column for which you want to calculate the average

result = calculate_column_average(file_path, column_name)

if isinstance(result, float):
    print(f"Average of values in column '{column_name}': {result:.2f}")
else:
    print(result)
```

# 5. A program that reads an Excel file and prints the data in a tabular format.

```python
import pandas as pd

def print_excel_data(file_path):
    try:
        # Read the Excel file into a pandas DataFrame
        df = pd.read_excel(file_path)

        # Print the DataFrame
        print("Data in Tabular Format:")
        print(df)

    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
file_path = 'sample_data.xlsx'  # Replace with the path to your Excel file
print_excel_data(file_path)
```
